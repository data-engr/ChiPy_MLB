```python
# Import packages
import pandas as pd
import MLB_CONFIG
from random import randint
from time import sleep


# Step 1:
# create stadium dictionary

# stadiums_dict = {'stadium':["LAA", "HOU","OAK","TOR","ATL","MIL","STL","CHC","ARI","LAD","SF","CLE","SEA","MIA","NYM","WSH","BAL","SD","PHI","PIT","TEX","TB","BOS","CIN","COL","KC","DET","MIN","CWS","NYY"]}      

        
# Step 2 : Download Data
def get_pitch_fx_data(stadiums_dict, date_str):
    pitch_fx_data_df = pd.DataFrame()
    for st in stadiums_dict["stadium"]:
        print(date_str+" : "+st)
        url = "https://baseballsavant.mlb.com/statcast_search/csv?all=true&hfPT=&hfAB=&hfBBT=&hfPR=&hfZ=&stadium={}&hfBBL=&hfNewZones=&hfGT=R|&hfC=&hfSea=2018|&hfSit=&player_type=pitcher&hfOuts=&opponent=&pitcher_throws=&batter_stands=&hfSA=&game_date_gt={}&game_date_lt={}&hfInfield=&team=&position=&hfOutfield=&hfRO=&home_road=&hfFlag=&hfPull=&metric_1=&hfInn=&min_pitches=0&min_results=0&group_by=name&sort_col=pitches&player_event_sort=h_launch_speed&sort_order=desc&min_pas=0&type=details&".format(st, date_str, date_str)
        data = pd.read_csv(url, na_values = MLB_CONFIG.bsavant_null_dict)
        pitch_fx_data_df = pitch_fx_data_df.append(data, ignore_index=True)
        sleep(randint(1,2))
        
    pitch_fx_data_df['batter'] = pd.to_numeric(pitch_fx_data_df.batter, errors='coerce')
    pitch_fx_data_df['pitcher'] = pd.to_numeric(pitch_fx_data_df.pitcher, errors='coerce')
    pitch_fx_data_df['zone'] = pd.to_numeric(pitch_fx_data_df.zone, errors='coerce')
    pitch_fx_data_df['hit_location'] = pd.to_numeric(pitch_fx_data_df.hit_location, errors='coerce')
    pitch_fx_data_df['balls'] = pd.to_numeric(pitch_fx_data_df.balls, errors='coerce')
    pitch_fx_data_df['strikes'] = pd.to_numeric(pitch_fx_data_df.strikes, errors='coerce')
    pitch_fx_data_df['game_year'] = pd.to_numeric(pitch_fx_data_df.game_year, errors='coerce')
    pitch_fx_data_df['outs_when_up'] = pd.to_numeric(pitch_fx_data_df.outs_when_up, errors='coerce')
    pitch_fx_data_df['inning'] = pd.to_numeric(pitch_fx_data_df.inning, errors='coerce')
    pitch_fx_data_df['fielder_2'] = pd.to_numeric(pitch_fx_data_df.fielder_2, errors='coerce')
    pitch_fx_data_df['release_spin_rate'] = pd.to_numeric(pitch_fx_data_df.release_spin_rate, errors='coerce')
    pitch_fx_data_df['game_pk'] = pd.to_numeric(pitch_fx_data_df.game_pk, errors='coerce')
    pitch_fx_data_df['pitcher.1'] = pd.to_numeric(pitch_fx_data_df['pitcher.1'], errors='coerce')
    pitch_fx_data_df['fielder_2.1'] = pd.to_numeric(pitch_fx_data_df['fielder_2.1'], errors='coerce')
    pitch_fx_data_df['fielder_3'] = pd.to_numeric(pitch_fx_data_df.fielder_3, errors='coerce')
    pitch_fx_data_df['fielder_4'] = pd.to_numeric(pitch_fx_data_df.fielder_4, errors='coerce')
    pitch_fx_data_df['fielder_5'] = pd.to_numeric(pitch_fx_data_df.fielder_5, errors='coerce')
    pitch_fx_data_df['fielder_6'] = pd.to_numeric(pitch_fx_data_df.fielder_6, errors='coerce')
    pitch_fx_data_df['fielder_7'] = pd.to_numeric(pitch_fx_data_df.fielder_7, errors='coerce')
    pitch_fx_data_df['fielder_8'] = pd.to_numeric(pitch_fx_data_df.fielder_8, errors='coerce')
    pitch_fx_data_df['fielder_9'] = pd.to_numeric(pitch_fx_data_df.fielder_9, errors='coerce')
    pitch_fx_data_df['at_bat_number'] = pd.to_numeric(pitch_fx_data_df.at_bat_number, errors='coerce')
    pitch_fx_data_df['pitch_number'] = pd.to_numeric(pitch_fx_data_df['pitch_number'], errors='coerce')
    pitch_fx_data_df['home_score'] = pd.to_numeric(pitch_fx_data_df.home_score, errors='coerce')
    pitch_fx_data_df['away_score'] = pd.to_numeric(pitch_fx_data_df.away_score, errors='coerce')
    pitch_fx_data_df['bat_score'] = pd.to_numeric(pitch_fx_data_df.bat_score, errors='coerce')
    pitch_fx_data_df['fld_score'] = pd.to_numeric(pitch_fx_data_df.fld_score, errors='coerce')
    pitch_fx_data_df['post_away_score'] = pd.to_numeric(pitch_fx_data_df.post_away_score, errors='coerce')
    pitch_fx_data_df['post_home_score'] = pd.to_numeric(pitch_fx_data_df.post_home_score, errors='coerce')
    pitch_fx_data_df['post_bat_score'] = pd.to_numeric(pitch_fx_data_df.post_bat_score, errors='coerce')
    pitch_fx_data_df['post_fld_score'] = pd.to_numeric(pitch_fx_data_df.post_fld_score, errors='coerce')
    
    
       
    return pitch_fx_data_df

df = get_pitch_fx_data(MLB_CONFIG.stadiums_dict, "2018-09-21")
df.info()
print(df)

```
